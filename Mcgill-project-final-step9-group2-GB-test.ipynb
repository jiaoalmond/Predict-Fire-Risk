```python
group2['DATE'] = group2['DATE'].astype('datetime64[ns]') 
```

## create new features: month and day

```python
group2['MONTH'] = pd.to_datetime(group2['DATE']).dt.month
group2['DAY'] = pd.to_datetime(group2['DATE']).dt.day
group2
```

```python
group2 = group2.round(decimals = 3)
```

```python
group2=group2.drop(columns = ['ARRONDISSEMENT', 'DATE','CATEGORIE'])
```

## Encoding related categorical variables with dummy varaibles

```python
# generate binary values using get_dummies
model2 = pd.get_dummies(group2, columns=['QUART'])
```

```python
model2
```

```python
# save model2
model2.to_csv('D:/MCGILL-DATA SCIENCE/YCBS-299-PRJ/data/jjycleaned/2015-2019/model2.csv', index=False)
```

### data spliting

```python
train, validate, test = np.split(model2.sample(frac=1), [int(.6*len(model2)), int(.8*len(model2))])
```

```python
train.info()
validate.info()
test.info()
```

```python
train['YEAR'].value_counts()
validate['YEAR'].value_counts()
```

```python
y_train = train['RISK_LVL']
X_train= train.drop(columns=['RISK_LVL'], inplace=False)
y_train
```

```python
X_train
```

```python
y_validate = validate['RISK_LVL']
X_validate= validate.drop(columns=['RISK_LVL'], inplace=False)
X_validate,y_validate 
```

```python
y_test = test['RISK_LVL']
X_test= test.drop(columns=['RISK_LVL'], inplace=False)
X_test,y_test
```

## Gradient Boosting Classifier

```python
from matplotlib import pyplot
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lr_list = [0.1,0.2,0.25,0.5]

for learning_rate in lr_list:
    gb_clf = GradientBoostingClassifier(n_estimators=12, learning_rate=learning_rate, max_features=2, max_depth=2, random_state=0)
    gb_clf.fit(X_validate, y_validate)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
    print("Accuracy score (validate): {0:.3f}".format(gb_clf.score(X_validate, y_validate)))
```

```python
predictions = gb_clf.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))

print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# get importance
importance = gb_clf.feature_importances_

# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
    
show =pd.Series(importance, index=X_train.columns).nlargest(15).plot(kind='barh')
pyplot.title("Feature Importance")
pyplot.show()
```
