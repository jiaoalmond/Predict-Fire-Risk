```python
path = "D:/MCGILL-DATA SCIENCE/YCBS-299-PRJ/data/jjycleaned/"
weather = path + "Weather_Data_Montreal_2015-2019.xlsx"
weather= pd.read_excel(weather)
weather.head()
```

```python
weather.info()
```

```python
def f(x):
    if (x > 8) and (x <= 16):
        return 'jour'
    elif (x > 16) and (x <= 24 ):
        return 'soir'
    elif (x > 0) and (x <= 8):
        return'nuit'
    elif (x == 0) :
        return'nuit'
```

```python
weather['DATE'] = pd.to_datetime(weather['date_fixed'])
weather['Time'] = pd.to_datetime(weather['Time'], format='%H:%M:%S')


weather['HOUR'] = pd.to_datetime(weather['Time']).dt.hour
weather['QUART'] = weather['HOUR'].apply(f)
weather.head()
```

```python
weather= weather[['DATE','QUART', 'Temp (°C)', 'Dew Point Temp (°C)', 'Rel Hum (%)', 'Wind Spd (km/h)']]
```

```python
weather
```

### Arithmetic Mean

```python
Temp_avg = weather.groupby(['DATE','QUART'])
```

```python
weather_sum =weather.groupby(['DATE','QUART']).sum()
```

```python
weather_sum
```

```python
print(weather_sum.columns)
```

```python
weather_sum=weather_sum.reset_index(level=['DATE', 'QUART'])
```

```python
weather_sum
```

```python
weather_sum1=weather_sum.rename(columns={"Temp (°C)": "Temp_sum", 'Dew Point Temp (°C)':'Dew_sum', 'Rel Hum (%)': 'Hum_sum', 'Wind Spd (km/h)':'Wind_sum'})
```

```python
weather_sum1
```

```python
weather_len = weather.groupby(['DATE','QUART']).size()
```

```python
weather_len=pd.DataFrame(weather_len)
```

```python
weather_len
```

```python
weather_len=weather_len.reset_index(level=['DATE', 'QUART'])
weather_len.rename(columns={0:'len'}, 
                 inplace=True)
```

```python
weather_len
```

```python
weather_sum1.info()
weather_len.info()
```

```python
weather_cal=pd.merge(weather_sum1, weather_len, on=['DATE', 'QUART'])
```

```python
weather_cal['Temp_avg']= weather_cal['Temp_sum']/weather_cal['len']
weather_cal['Dew_avg']= weather_cal['Dew_sum']/weather_cal['len']
weather_cal['Hum_avg']= weather_cal['Hum_sum']/weather_cal['len']
weather_cal['Wind_avg']= weather_cal['Wind_sum']/weather_cal['len']
weather_cal.round(decimals=2)
```

```python
weather_cal = weather_cal.round(decimals=2)
weather_cal = weather_cal[['DATE','QUART','Temp_avg','Dew_avg','Hum_avg','Wind_avg']]
```
